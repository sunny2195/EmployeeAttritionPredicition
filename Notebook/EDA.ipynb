```python
import pandas as pd
import numpy as np 
```

```python
df= pd.read_csv("D:\Projects\Employee Attiriton Prediction\Data\EmployeeData.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['EmployeeCount', 'StandardHours','Over18','EmployeeNumber'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
binary_and_target_cols = ['Attrition','Gender','OverTime']
le = LabelEncoder()
for col in binary_and_target_cols:
    if col in df.columns:
        df[col] = le.fit_transform(df[col])
```

```python
one_hot_cols = [
    'BusinessTravel',
    'Department',
    'EducationField',
    'JobRole',
    'MaritalStatus'
]
```

```python
df = pd.get_dummies(df, columns=one_hot_cols, drop_first=True,dtype=int)
```

```python
df.to_csv("D:\Projects\Employee Attiriton Prediction\Data\EmployeeData_Processed.csv", index=False)
```
